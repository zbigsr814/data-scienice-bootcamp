```python
def fetch_financal_data(company = 'AMZIN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

apple = fetch_financal_data('AAPL')
bp = fetch_financal_data('BP')
google = fetch_financal_data('GOOGL')
uber = fetch_financal_data('UBER')
```

```python
bp.head()
```

Preprocessing

```python
apple.columns = ['apple_' + col.lower() for col in apple.columns]
bp.columns = ['bp_' + col.lower() for col in bp.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
bp
```

```python
import pandas as pd
```

```python
df = pd.concat(objs=[apple, bp, google, uber], axis=1)
df.head()
```
