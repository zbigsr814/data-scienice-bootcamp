```python
df = fetch_financal_data('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Average'] = (df['Open']+df['Close'])/2
df.head()
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
df[['Close']].shift(1)
```

```python
df['Close_shift'] = df['Close'].shift(1)
df.head()
```

```python
df['Daily_change'] = df['Close'] / df['Close_shift'] - 1
```
