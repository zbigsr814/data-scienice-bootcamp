```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

Wykres liniowy

```python
_ = plt.plot([0,1],[0,3])
```

```python
_ = plt.plot([0,3,1], [0,1,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-5,5,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

Tworzenie subplotów

```python
fig = plt.figure(figsize=(12,8))
plt.subplot(121)
plt.plot(x,x_sin, 'g.')
plt.subplot(122)
plt.plot(x,x_cos, 'r--')
```

Wykresy słupkowe

```python
plt.barh(y=['a','b','c','d'], width=[1.2,3.3,0.3,5.5])
```

```python
x1 = np.random.randn(300)
y1 = np.random.randn(300)
x3 = np.random.randn(300)*50
plt.scatter(x=x1, y=y1, s=x3, alpha=0.5)
```
