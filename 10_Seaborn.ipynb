```python
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

Import danych

```python
df = sns.load_dataset('tips')
df.head(5)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time')
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', row='time')
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='smoker', row='time')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```
